```python
def read_out_file(folder):
    d = {}
    for i in range(3,21):
        d['omega_{}'.format(i)] = pd.read_csv("wyniki/{}/Omega {}pi2/sim.out".format(folder,i), sep='\t', header=0, index_col=False)
        
    return d

def read_out_file2(folder, fr, to, n):
    d = {}
    for i in np.linspace(fr, to, n):
        d['omega_{}'.format(i)] = pd.read_csv("wyniki/{}/sim{}.out".format(folder,i), sep='\t', header=0, index_col=False)
        
    return d
```

```python
#d_n4 = read_out_file('K2n4_v2')
#d_n1 = read_out_file('K2n1_v2')
d_n1 = read_out_file2('K10n1_domiar', 2, 4, 20)
d_n9 = read_out_file2('K10n9_domiar', 18, 25, 30)

#d_n4_v2 = read_out_file2('K2n4_domiar', 6, 10, 20)
d_n4 = read_out_file2('K10n4_domiar', 7, 10, 20)

#d_n4 = dict(list(d_n4.items()) + list(d_n4_v2.items()))
```

```python
def find_max_omega(dic):
    m = list(dic.keys())[0]
    for d in dic.keys():
        if dic[m]['epsilon_out'].max() < dic[d]['epsilon_out'].max():
            m = d
    return m

def prepare_omega_list(dic):
    l = []
    for d in dic.keys():
        l.append(dic[d]['epsilon_out'].max())
        
    return l
```

```python
n4_omega_list = prepare_omega_list(d_n4)
n1_omega_list = prepare_omega_list(d_n1)
n9_omega_list = prepare_omega_list(d_n9)

n4_omega_list_K10 = prepare_omega_list(d_n4_K10)
```
