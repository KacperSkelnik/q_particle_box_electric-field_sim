```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
```

```python
fig, ax = plt.subplots()

ax.set_xlim((0, 1))
ax.set_ylim((0, 3.5))

line, = ax.plot([], [], lw=2)
```

```python
def init():
    line.set_data([], [])
    return (line,)
```

```python
def animate(i):
    path = 'wyniki/K 1_5 n 4/Omega 8pi2/sim.dat'
    
    x = np.linspace(0, 1, 51)
    
    f = open(path, 'r') 
    y = []
    for ln in f.readlines():
        y_tmp = [float(i) for i in ln.split("\t")]
        y.append(y_tmp)

    line.set_data(x, y[i])
    return (line,)
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=800, interval=10, blit=True)
```

```python
HTML(anim.to_html5_video())
```
