```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
```

```python
fig, ax = plt.subplots()

fig.set_figheight(9)
fig.set_figwidth(16)

ax.set_xlim((0, 1))
ax.set_ylim((0, 5))

line, = ax.plot([], [], lw=2)
```

```python
def init():
    line.set_data([], [])
    return (line,)
```

```python
path = 'wyniki/K10n1_domiar/sim3.789473684210526.dat'

def animate(i):
    x = np.linspace(0, 1, 51)
    
    f = open(path, 'r') 
    y = []
    for ln in f.readlines():
        y_tmp = [float(i) for i in ln.split("\t")]
        y.append(y_tmp)

    line.set_data(x, y[i])
    return (line,)
```

```python
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=3000, interval=10, blit=True)
```

```python
HTML(anim.to_html5_video())
```
